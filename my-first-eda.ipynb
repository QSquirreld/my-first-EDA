## Анализ 1(Где покупать недвижимость)

**Сгруппируем данные из фрейма по континентам и отсортируем их в порядке густонаселённости на 2022 год, чтобы оценить темпы роста и найти самый быстрорастущий:**

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population",ascending=False)
```

* **Самый быстрорастущий континент - Африка, пока он на третьей позиции, но с сохранением тенденций сможет обогнать Южную Америку, до Азии вряд ли кто-то дотянется**
* **В противоположность Азии, Южная Америка имеет наименьшую плотность населения(вероятно из-за джунглей), однако занимает второе место по численности. Значит там ещё полно места).**

**Проверим в какой стране Южной Америки лучше накупить земли на будущее:**
<br>*также уберём всякие острова и страны менее 1млн км площади*

```python
df[(df['Continent'].str.contains('South America')) & (df['Area (km²)'] > 1000000)].sort_values(by='Density (per km²)')
```

**Вывод: Скупаем дома в Боливии, несмотря на политику, преступность и прочие "мелочи"**

## Анализ 2(Где не стоит строить "единорога", по-крайней мере Хартманн не стал бы(а может и стал бы, он же умный))

**Теперь хочется наглядно посмотреть на рост населения в странах Азии**

```python
## создадим субфрейм для стран Азии, для лучшей читабельности
df_asia = df.loc[df['Continent'] == 'Asia'] # df[df['Continent'] == 'Asia'] - тоже подошло бы
df_asia
```

**Оставим только интересующие нас столбцы**

```python
df2 = df_asia[['Country'] + list(df.columns[12:4:-1])].sort_values(by="2022 Population",ascending=False)
df2.set_index('Country', inplace = True)
df2
```

```python
df2.iloc[2:7].transpose().plot(figsize=(15, 5))
```

**Уберём из выборки больших ребят в лице Китая и Индии и заметим, что Япония теряет население**

```python
df2.loc['Japan']
```

**Как видим с 2010го население уменьшается**

```python
population_difference = df2.loc['Japan', '2022 Population'] - df2.loc['Japan', '2010 Population']
population_difference
```
